5주차 과제

국내 1인당 소득, 소비자 물가 지수, 가계소비지출 간에 어떤 관계가 있는지 알아보았습니다.

3개의 데이터는 kosis.kr에서 2009년부터 2019년까지 추출하였습니다.(가계지출이 2009년부터 2019년의 데이터 밖에 존재하지 않았습니다.)

데이터의 수가 짧아 3개의 데이터 간에 연관관계가 있는지 확실하게 말하기는 어렵지만,

11년동안 소득이 꾸준히 증가하였고, 그에 따라 물가지수도 증가하였고, 가계소비역시 증가하였다는 것을 볼 수 있었습니다.

```python
# 구글 코랩 한글 관련
!apt-get install fonts-nanum -qq > /dev/null
!fc-cache -fv
import matplotlib as mpl
mpl.font_manager._rebuild()
findfont = mpl.font_manager.fontManager.findfont
mpl.font_manager.findfont = findfont
mpl.backends.backend_agg.findfont = findfont

# 국내 1인당 소득 csv 불러오기
import csv
data1 = csv.reader(open('income.csv', encoding = 'cp949'))
next(data1)
# 국내 1인당 소득 리스트 만들기 및 초드학교 취학생 수 변화를 알고 싶은 동 물어보기
income = []
# 국내 1인당 소득 데이터를 리스트에 넣기
for row in data1 :
  for i in range(1, 12) :
    income.append(int(row[i]))
  break

# 소비자물가지수 csv 불러오기
data2 = csv.reader(open('consumer.csv', encoding = 'cp949'))
next(data2)
# 소비자물가지수 리스트 만들기 및 초드학교 취학생 수 변화를 알고 싶은 동 물어보기
consumer = []
# 소비자물가지수 데이터를 리스트에 넣기
for row in data2 :
  for i in range(1, 12) :
    consumer.append(float(row[i]))
  break

# 가계소비지출 csv 불러오기
data3 = csv.reader(open('expenditure.csv', encoding = 'cp949'))
next(data3)
# 가계소비지출 리스트 만들기 및 초드학교 취학생 수 변화를 알고 싶은 동 물어보기
expenditure = []
# 가계소비지출 데이터를 리스트에 넣기
for row in data3 :
  for i in range(1, 12) :
    expenditure.append(float(row[i]))
  break

# 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize=(10,10))
# 국내 1인당 소득 그래프 그리기
plt.subplot(311)
plt.plot(income)
plt.xticks(range(11),[2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])
plt.title('국내 1인당 소득 (단위:천원)')
plt.xlabel('연도')
# 소비자물가지수 그래프 그리기
plt.subplot(312)
plt.plot(consumer)
plt.xticks(range(11),[2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])
plt.title('소비자물가지수 (2020년=100기준)')
plt.xlabel('연도')
# 가계소비지출 그래프 그리기
plt.subplot(313)
plt.plot(expenditure)
plt.xticks(range(11),[2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])
plt.title('가계소비지출 (단위:천원)')
plt.xlabel('연도')
plt.tight_layout()
plt.show()
```
